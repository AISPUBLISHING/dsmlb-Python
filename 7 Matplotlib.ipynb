```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
a = np.linspace(0, 10, 100)
#print(a)

fig = plt.figure()
plt.plot(a, np.sin(a), '--')
plt.plot(a, np.cos(a), '-');
```

```python
a = np.linspace(0, 10, 100)
plt.figure()

#creates the 1st of two panels and set current axis

plt.subplot(2, 1, 1) #(rows, columns, panel number)
plt.plot(a, np.sin(a), '-')


plt.subplot(2, 1, 2)
plt.plot(a, np.cos(a), '--')
```

```python
plt.plot(a, np.sin(a - 0), color='black')        # specify color by name
plt.plot(a, np.sin(a - 1), color='b')           # short color code (rgbcmyk)
plt.plot(a, np.sin(a - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(a, np.sin(a - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(a, np.sin(a - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(a, np.sin(a - 5), color='chartreuse') # all HTML color names supported
```

```python
plt.plot(a, np.sin(a))

plt.xlim(-3, 11)
plt.ylim(-1.0, 1.0)
```

```python
plt.plot(a, np.sin(a))
plt.title("Sine Curve...")
plt.xlabel("x -> ")
plt.ylabel("sin(x) -> ")
```
