```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('data/banking.csv', header=0)
data.shape
```

```python
data.head()
```

```python
class_le = LabelEncoder()
for column in data[["job", "marital", "education", "default", "housing", "loan", "contact", "month", "day_of_week", "poutcome"]].columns:
    data[column] = class_le.fit_transform(data[column].values)
    
```

```python
data.head()
```

```python
X = data.loc[:, data.columns != 'y']
y = data.loc[:, data.columns == 'y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
lrc = LogisticRegression()
lrc.fit(X_train, y_train)
```

```python
print('Accuracy of logistic regression classifier on test set:', lrc.score(X_test, y_test))
```
