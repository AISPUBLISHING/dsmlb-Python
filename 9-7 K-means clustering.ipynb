```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
a= -3 * np.random.rand(200,2)
a1 = 1 + 3 * np.random.rand(100,2)
a[100:200, :] = a1

plt.scatter(a[ : , 0], a[ :, 1], s = 100, c = 'r')
plt.show()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(a)
```

```python
Kmean.cluster_centers_
```

```python
plt.scatter(a[ : , 0], a[ : , 1], s =100, c='r')
plt.scatter(Kmean.cluster_centers_[0][0], Kmean.cluster_centers_[0][1], s=200, c='y', marker='s')
plt.scatter(Kmean.cluster_centers_[1][0], Kmean.cluster_centers_[1][1], s=200, c='b', marker='s')
plt.show()
```
